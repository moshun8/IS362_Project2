# Income Data

```python
import pandas as pd
import numpy as np

income = pd.read_csv('https://raw.githubusercontent.com/moshun8/IS362_Project2/master/Project2_personalIncome.csv')
#income = pd.read_csv('/Users/Maureen/Desktop/GitHub/IS362_Project2/Project2_personalIncome.csv')
income.head()
```

### Get rid of the "Line" column and rename others since the headers were numbers

I'm working under the assumption that the original 2014, 2015, and 2016 columns were meant to be Q0.

```python
column_names = {'A': 'type', 
                '2014': 'y14q0',
                '2014.1': 'y14q1', 
                '2014.2': 'y14q2', 
                '2014.3': 'y14q3',
                '2015': 'y15q0', 
                '2015.1': 'y15q1',
                '2015.2': 'y15q2',
                '2015.3': 'y15q3',
                '2016': 'y16q0',
                '2016.1': 'y16q1',
                '2016.2': 'y16q2',
                '2016.3': 'y16q3',}

income = income.drop('Line',axis=1).rename(columns=column_names)
income.head()
```

### Make new columns with the mean for each year

```python
income['all2014'] = income[['y14q0','y14q1','y14q2','y14q3']].mean(axis=1)
income['all2015'] = income[['y15q0','y15q1','y15q2','y15q3']].mean(axis=1)
income['all2016'] = income[['y16q0','y16q1','y16q2','y16q3']].mean(axis=1)
```

### Next Steps:
1. Use melt to switch to long format
2. Sort and assign new index
3. Calculate percent change for each row, change every 15th value NaN

```python
income_melt = pd.melt(income,
                ["type"],
                var_name="year",
                value_name="amount")

income_melt = (income_melt
               .sort_values(by=['type',"year"])
               .set_index(['type','year']))

income_melt['%change'] = income_melt['amount'].pct_change()
income_melt['%change'].iloc[0::15]=np.nan
income_melt['%change'].iloc[3::15]=np.nan
income_melt.head(30)
```
